```python
##读取数据
comfirm_file_path = 'COVID-19/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv'
import pandas as pd

country_mapping = {
    'Mainland China': 'China',
    'UK': 'United Kingdom',
    'US': 'United States'
}

df = pd.read_csv(comfirm_file_path)
cols = list(df.columns[4:])
print("疫情时间：", cols)

# 计算每个国家的数据
group_df = df.groupby(by='Country/Region',as_index=False)[cols].sum()
```

```python
# 显示地图
from pyecharts.faker import Faker
from pyecharts import options as opts
from pyecharts.charts import Map
from pyecharts.charts import Timeline
## 现在各国的
t = Timeline()
for date in cols:
    subdf = group_df[group_df[date] != 0]
    Country = subdf['Country/Region'].apply(lambda x: country_mapping[x] if x in country_mapping.keys() else x)
    confirmed = subdf[date]
    
    c = (
        Map()
        .add("Covid-19", [list(z) for z in zip(Country, confirmed)], "world")
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(
            title_opts=opts.TitleOpts(title="世界疫情情况"+date),
            visualmap_opts=opts.VisualMapOpts(min_=min(confirmed), max_=max(confirmed)/2),
        )
    )
    t.add(c,date)

t.render_notebook()
```
